Write a program to download the data from the link given below and then read the data and convert the into
the proper structure and return it as a CSV file.
Link - https://data.nasa.gov/resource/y77d-th95.json
Note - Write code comments wherever needed for code understanding.

Excepted Output Data Attributes
● Name of Earth Meteorite - string id - ID of Earth
● Meteorite - int nametype - string recclass - string
● mass - Mass of Earth Meteorite - float year - Year at which Earth
● Meteorite was hit - datetime format reclat - float recclong - float
● point coordinates - list of int


```python
import pandas as pd
```

```python
import urllib

from urllib.request import urlopen

 

# import json

import json
from pandas.io.json import json_normalize

# store the URL in url as

# parameter for urlopen

url = "https://data.nasa.gov/resource/y77d-th95.json"

 

# store the response of URL

response = urlopen(url)

 

# storing the JSON response

# from url in data

data_json = json.loads(response.read())

 

df=json_normalize(data_json)
```

```python
df.year
```

```python
df["year"]=df.year.str.split("-").str[0]
```

```python
# ● Name of Earth Meteorite - string 
# id - ID of Earth Meteorite - int
# nametype - string 
# recclass - string
# ● mass - Mass of Earth Meteorite - float 
# year - Year at which Earth Meteorite was hit  datetime format int
#  reclat - float 
# recclong - float
# ● point coordinates - list of int
```

```python
type(df["id"])
```

```python
df["id"].dtype
```

```python
df["id"]=df["id"].astype(int)
```

```python
df["id"].dtype
```

```python
df["mass"]=df["mass"].astype(float)
```

```python
df["year"]=df["year"].astype(int)
```

```python
df["year"].isnull().sum()
```

```python
df["year"].dropna(how="any",axis=0,inplace=True)
df["year"]=df["year"].astype(float)
```

```python
df["reclat"]=df["reclat"].astype(float)
```

```python
df["reclong"]=df["reclong"].astype(float)
```

```python
df["reclong"].dtype
```

```python
df["geolocation.coordinates"].dtype
```

```python
df["reclong"]=df["reclong"].astype(int)
```

```python
df['geolocation.coordinates'] = df['geolocation.coordinates'].astype(str)
```

```python
df['geolocation.coordinates'] = df['geolocation.coordinates'].replace('.','')
```

```python
type(df['geolocation.coordinates'][0])
```

```python
import ast

df['geolocation.coordinates'] = df['geolocation.coordinates'].apply(ast.literal_eval)
df['geolocation.coordinates'] = df['geolocation.coordinates'].apply(lambda x: [int(i) for i in x])
```

```python
df
```

```python
df.to_csv("Nasa.csv",index=False,header=True)
```

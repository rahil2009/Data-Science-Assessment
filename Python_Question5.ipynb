```python
# Excepted Output Data Attributes -
# ● id - int url - string
# ● name - string season
# ● - int number - int
# ● type - string airdate -
# ● date format airtime -
# ● 12-hour time format
# ● runtime - float
# ● average rating - float
# ● summary - string
# ● without html tags
# ● medium image link - string
# ● Original image link - string
```

```python
import urllib
from urllib.request import urlopen
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
link = 'http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes'
response = requests.get(link)        
```

```python
if response.status_code == 200:  # Check if the request was successful
    data = response.json()  # Extract the JSON data from the response
    # Process the data as per your requirements
else:
    print("Error:", response.status_code)
```

```python
data
```

```python
x = json_normalize(data)
```

```python
x
```

```python
df1 = data['_embedded']['episodes']
```

```python
df1
```

```python
df = pd.DataFrame(df1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['airdate'] = pd.to_datetime(df['airdate'])
```

```python
df['airtime'] = pd.to_datetime(df['airtime']).dt.strftime('%I:%M %p')
```

```python
df['airtime'].unique()
```

```python
df['runtime'] = df['runtime'].astype(float)
```

```python
lst=[]
for i in df['rating']:
    lst.append(list(i.values())[0])
```

```python
df['rating']=lst
```

```python
df['summary'][0]
```

```python
from bs4 import BeautifulSoup

def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

html_string = "<p>This is an <b>example</b> HTML string.</p>"
clean_text = remove_html_tags(html_string)
print(clean_text)
```

```python
df['summary'] = df['summary'].apply(remove_html_tags)
```

```python
lst = []
lst2 = []

for i in df['image']:
    lst.append(i['medium'])
for i in df['image']:
    lst2.append(i['original'])
```

```python
df['Medium Image link'] = lst
df['original Image link'] = lst2
```

```python
df
```

```python
df.to_csv('tvmaze.csv', index=False, header = True )
```
